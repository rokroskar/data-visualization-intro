# Importing and displaying some X-ray data

```python
# load numerical analysis library.
import numpy as np

# load data handling libraries.
import pandas as pd

# load plotting libraries.
import matplotlib
from matplotlib import pyplot as plt

# file handling libraries.
import os
from pathlib import Path
```

```python
# path to data file
fpath_data = "./data/Federica_Bertolotti/Fh.gr"
# fpath_data = str(Path(r"C:\Users\bosi\Dropbox\ETH\boards\SCS_Materials_Chemistry\Symposium\MatChem25\school\data_visualization\examples\Federica_Bertolotti\data\Fh.gr"))
```

```python
# load data from file
# the data block are only the lines after the string "#### start data"
# of this block, the lines starting with # are comments and should be ignored.# the file is a space-separated file with a header. The header is marked by lines starting with #
data = pd.read_csv(fpath_data, skiprows=25, comment="#", delimiter=" ", header=None)
data
```

```python
# plot the data
fig, ax = plt.subplots(figsize=(8, 6))
ax.plot(data[0], data[1], label="Fh", color="blue", linewidth=1)
ax.set_xlabel("r (Å)")
ax.set_ylabel(r"G (Å$^{2}$)")
ax.legend()

# save it to a file in the current working directory
fpath_out = os.path.join(os.getcwd(), "Fh_plot.png")
fig.savefig(fpath_out, dpi=300, bbox_inches="tight")
```
